```python
import requests
import os

# Configuration
taxon_id = 118589  # Conger conger
max_pages = 5      # Modifier pour plus d'images (100 par page)
save_dir = "images_congre"
os.makedirs(save_dir, exist_ok=True)

for page in range(1, max_pages + 1):
    url = "https://api.inaturalist.org/v1/observations"
    params = {
        "taxon_id": taxon_id,
        "quality_grade": "research",
        "has[]": "photos",
        "per_page": 100,
        "page": page
    }

    print(f"Récupération page {page}...")
    response = requests.get(url, params=params)
    data = response.json()

    for obs in data['results']:
        obs_id = obs['id']
        if obs['photos']:
            photo_url = obs['photos'][0]['url'].replace('square', 'original')
            filename = f"{obs_id}.jpg"
            try:
                img_data = requests.get(photo_url).content
                with open(os.path.join(save_dir, filename), 'wb') as f:
                    f.write(img_data)
                print(f"Image sauvegardée : {filename}")
            except Exception as e:
                print(f"Erreur avec {filename}: {e}")
```
